```python
y=recovered_data[["end"]]
x=recovered_data[['tranc', 'price', 'high', 'low']]
print(x)
print(y)
```

```python
def create_dataset(X, Y, time_step, step):
    Xs, Ys = [], []

    for i in range(0, len(X) - time_step, step):
        v = X.iloc[i : (i + time_step)].values
        labels = Y.iloc[i + time_step]

        Xs.append(v)
        Ys.append(labels)

    return np.array(Xs), np.array(Ys)
```

```python
XX, yy = create_dataset(x, y, 10, 1)

XX.shape
```

```python
XX[::3]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(XX, yy, test_size=0.3, random_state=421, shuffle=False)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_test
```

```python
y_test
```

```python

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout

model = Sequential()
model.add(LSTM(units = 64, activation='tanh', return_sequences = True, input_shape = (10, 4)))
model.add(Dropout(0.1))
model.add(LSTM(units = 32, activation='tanh'))
model.add(Dropout(0.1))
model.add(Dense(units = 1))
model.summary()


model.summary()
model.compile(loss='mse', optimizer='adam')
history = model.fit(X_train, y_train, epochs=1000, batch_size=32, validation_split=0.3, shuffle=False)
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']

x = np.arange(0, len(history.history['loss']))

plt.plot(x, loss, 'y', label='loss')
plt.plot(x, val_loss, 'r', label='val loss')
plt.title('Losses')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
plt.show()
```

```python
y_pred = model.predict(X_test)
y_pred.shape

plt.plot(y_test, color='red', label='real stock price')
plt.plot(y_pred, color='blue', label='predicted stock price')
plt.title('stock price prediction')
plt.xlabel('time')
plt.ylabel('stock price')
plt.legend()
plt.show()
```
