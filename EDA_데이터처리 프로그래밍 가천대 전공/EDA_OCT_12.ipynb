```python
import pandas as pd
df = pd.read_csv("04_data1.csv", index_col = 0)
```

```python
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
index = np.arange(7)
plt.figure(figsize=(10,4))
data6 = df.loc['Seoul':'Ulsan', '2017_common':'2020_common']
for year in range(2017,2021):
    year = str(year)+'_common'
    chartdata = data6[year]
    plt.barh(index, chartdata, label=year)
plt.xlim(120,310)
plt.ylabel('Area')
plt.xlabel('Micrometer')
plt.yticks(index, ['Seoul', 'Busan', 'Daegu', 'Incheon', 'Gwangju', 'Daejeon', 'U1san'])
plt.title('2017~2020 Fine Dust(pm10) Common Group Barh Graph')
plt.legend()
plt.show()
```

```python
features=df2.copy().head(100)
features
```

```python
features['Age'].fillna(features['Age'].median()).tail()
features=features.dropna()
```

```python


features = features.drop(columns=['Name', 'Ticket','Cabin'])
features['Sex'] = features['Sex'].replace({'male': 0, 'female': 1})
features['Embarked'] = features['Embarked'].map( {'C': 0, 'Q': 1, 'S':2} ).astype(int)
features
```

```python
plt.figure(figsize=(10,10))
plt.hist(features['Age'],bins=9,label='Age',alpha=0.5)
plt.hist(fare,bins=9,label='Fare',alpha=0.5)
plt.xlabel(' ')
plt.ylabel(' ')
plt.title('title~')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.hist(features['Age'],bins=9,label='Fare')
plt.xlabel('age')
plt.ylabel('year')
plt.title('title~')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.boxplot([features['Fare'],features['Age']],labels=['Fare','Age'])
plt.show()
```
