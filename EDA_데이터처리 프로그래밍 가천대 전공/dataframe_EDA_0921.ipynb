```python
import pandas as pd
student_card = pd.DataFrame({'ID':[0, 2, 1,9,8,7,6,5,4,3],
                             '학년':['1', '2', '3', '3', '3', '2', '1', '3', '1', '3'],
                             '수학':['A+', 'C+', 'B+', 'C+', 'C+', 'B+', 'B+', 'A+', 'B+', 'A+'],
                             '영어':['B+', 'A+', 'D+', 'C+', 'D+', 'A+', 'B+', 'B+', 'B+', 'D+'],
                             '과학':['B+', 'B+', 'A+', 'C+', 'D+', 'A+', 'C+', 'D+', 'D+', 'A+']})

student_card
```

```python
# 인코딩 딕셔너리 정의
grade_encoding = {
    'A+': 50,
    'B+': 40,
    'C+': 30,
    'D+': 20,
    'F': 10
}

# 각 성적 칼럼에 대해 인코딩 적용
for column in ['수학', '영어', '과학']:
    student_card[column] = student_card[column].map(grade_encoding)

student_card
```

```python
student_card['평균']=(student_card['수학']+student_card['영어']+student_card['과학'])/3
```
